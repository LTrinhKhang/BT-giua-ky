```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
folder = '/content/drive/MyDrive/train_set/'

for i in range(40):
    plt.subplot(10,6,i+1)
    filename = folder + 'Cuong ('+ str(i+1)+').jpg'
    img = imread(filename)
    plt.imshow(img)
plt.show()


from numpy import asarray
from os import listdir
from numpy import asarray,save
from keras.utils import img_to_array,load_img
folder = '/content/drive/MyDrive/train_set/'
photos, labels=list (),list()
for file in listdir(folder):
  output=0.0
  if file.startswith ('Cuong'):
    output = 1.0
  if file.startswith ('Danh'):
    output = 2.0
  if file.startswith ('Dat'):
    output = 3.0
  if file.startswith ('Dien'):
    output = 4.0
  if file.startswith ('Khang '):
    output = 5.0
  if file.startswith ('Luan'):
    output = 6.0
  if file.startswith ('Phat'):
    output = 7.0
  if file.startswith ('Phuc'):
    output = 9.0
  if file.startswith ('Quoc'):
    output = 10.0
  if file.startswith ('Tan'):
    output = 11.0
  if file.startswith ('Thuan'):
    output = 12.0
  if file.startswith ('Viet'):
    output = 8.0
  img =load_img(folder+ file,target_size=(30,40))
  photo = img_to_array(img)
  photos.append(photo)
  labels.append(output)
photos =asarray(photos)
labels =asarray(labels)
print(photos.shape,labels.shape)
save('face_photo.npy',photos)
save('face_labels.npy',labels)


import numpy as np
x_train = np.load('/content/face_photo.npy')
y_train = np.load('/content/face_labels.npy')
print(x_train.shape)
print(y_train.shape)
x=x_train
y=y_train

from keras.utils import to_categorical
x_train = x_train.astype('float32')/255
y_train = to_categorical(y_train,100)

from keras.backend import flatten
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D,Normalization,LeakyReLU
from keras.optimizers import Adam
#32 lần
model = Sequential()
model.add(Conv2D(32,kernel_size = (3,3),activation = 'relu',input_shape=(30,40,3),padding='Same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#64 lần
model.add(Conv2D(64,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#128 lần
model.add(Conv2D(128,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128,activation='linear'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(100,activation='softmax'))
from keras.losses import categorical_crossentropy
model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
model.summary()
train = model.fit(x_train,y_train,epochs=128,batch_size=64,verbose=1)
```

```python
model.save('face_rec.h5')
```
