```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
folder = '/content/drive/MyDrive/Food/'
for i in range(80):
  plt.subplot(10,8,i+1)
  filename=folder+'Pho ('+str(i+1)+').jpg'
  img=imread(filename)
  plt.imshow(img)
plt.imshow(img)
```

```python
from numpy import asarray
from os import listdir
from numpy import asarray,save
from keras.utils import img_to_array,load_img
folder = '/content/drive/MyDrive/Food/'
photos, labels=list (),list()
for file in listdir(folder):
  output=0.0
  if file.startswith ('Hu tieu'):
    output=1.0
  if file.startswith ('Com tam'):
    output=2.0
  if file.startswith ('Goi cuon'):
    output = 3.0
  if file.startswith ('Banh beo'):
    output = 4.0
  if file.startswith ('Nem chua'):
    output = 5.0
  if file.startswith ('Pho'):
    output = 6.0
  if file.startswith ('Banh gio'):
    output = 7.0
  if file.startswith ('Canh chua'):
    output = 8.0
  if file.startswith ('Chao long'):
    output = 9.0
  if file.startswith ('Banh mi'):
    output = 10.0
  img =load_img(folder+ file,target_size=(30,40))
  photo = img_to_array(img)
  photos.append(photo)
  labels.append(output)
photos =asarray(photos)
labels =asarray(labels)
print(photos.shape,labels.shape)
save('Food_photo.npy',photos)
save('Food_labels.npy',labels)
```

```python
import numpy as np
x_train = np.load('/content/Food_photo.npy')
y_train = np.load('/content/Food_labels.npy')
print(x_train.shape)
print(y_train.shape)
x=x_train
y=y_train
from keras.utils import to_categorical
x_train = x_train.astype('float32')/255
y_train = to_categorical(y_train,100)
```

```python
from keras.backend import flatten
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D,Normalization,LeakyReLU
from keras.optimizers import Adam
#32 ln
model = Sequential()
model.add(Conv2D(32,kernel_size = (3,3),activation = 'relu',input_shape=(30,40,3),padding='Same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#64 ln
model.add(Conv2D(64,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#128 ln
model.add(Conv2D(128,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#256 ln
model.add(Conv2D(256,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128,activation='linear'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(100,activation='softmax'))
from keras.losses import categorical_crossentropy
model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
model.summary()
train = model.fit(x_train,y_train,epochs=128,batch_size=64,verbose=1)
```

```python
import matplotlib.pyplot as plt
from keras.utils import load_img
from keras.utils.image_utils import img_to_array
import numpy as np
vat = {1: 'Hu tieu n',2:'Com tam n', 3:'Goi cuon n', 4:'Banh beo n',5:'Nem chua n',6:'Pho n',7:'Banh gio n',9:'Chao long n',8:'Canh chua n',10:'Banh mi n'}
img = load_img("/content/drive/MyDrive/test/Banh mi/1 (7).jpg",target_size=(30,40))
plt.imshow(img)
img = img_to_array(img)
img=img.reshape(1,30,40,3)
img = img.astype('float32')
img =img/255
result  = np.argmax(model.predict(img),axis=1)
vat[result[0]]
```
